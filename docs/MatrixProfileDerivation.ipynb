# Computing the Z-Normalized Euclidean Distance from Dot Products

In the [Matrix Profile I](https://www.cs.ucr.edu/~eamonn/STOMP_GPU_final_submission_camera_ready.pdf) and [Matrix Profile II](https://www.cs.ucr.edu/~eamonn/PID4481997_extend_Matrix%20Profile_I.pdf) papers, the Z-normalized Euclidean distance between a query subsequence, $Q=(q_0, q_1, \ldots, q_{m-1})$, and the $i^{th}$ subsequence, $T_{i,m}=(t_i, t_{i+1}, t_{i+2}, \ldots, t_{i+m-1})$, with window size, $m$, in the time series, $T$, can be computed following:

$ 
\begin{align}
    D(Q, T_{i,m}) &=
        \sqrt{
            \sum \limits _{0 \leq {j} \lt m}
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    - 
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)^2
        }
        \\
    &=
        \sqrt{
            \sum \limits _{0 \leq {j} \lt m}
                \left[
                    \left(
                        \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    \right)
                    -
                    2
                    \left(
                        \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    \right)
                    \left(
                        \frac{q_j-\mu_Q}{\sigma_Q}
                    \right)
                    +
                    \left(
                        \frac{q_j-\mu_Q}{\sigma_Q}
                    \right)^2
                \right]
        }
        \\
    &=
        \sqrt{
            \sum \limits _{0 \leq {j} \lt m}
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                \right)^2
                -
            \sum \limits _{0 \leq {j} \lt m}
                2
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                \right)
                \left(
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)
                +
            \sum \limits _{0 \leq {j} \lt m}
                \left(
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)^2
        }
        \\
    &=
        \sqrt{
            m
            -
            \sum \limits _{0 \leq {j} \lt m}
                2
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                \right)
                \left(
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)
            +
            m
        }
        \\
    &=
        \sqrt{
            2m
            -
            2
            \sum \limits _{0 \leq {j} \lt m}
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                \right)
                \left(
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{1}{m}
                \sum \limits _{0 \leq {j} \lt m}
                    \left(
                        \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    \right)
                    \left(
                        \frac{q_j-\mu_Q}{\sigma_Q}
                    \right)
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \sum \limits _{0 \leq {j} \lt m}
                    \frac{
                        \left(
                            t_{i+j}-M_{T_{i,m}}
                        \right)
                        \left(
                            q_j-\mu_Q
                        \right)
                    }{
                        m \sigma_Q \Sigma_{T_{i,m}}
                    }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \sum \limits _{0 \leq {j} \lt m}
                    \frac{
                        t_{i+j}q_j
                        -t_{i+j}\mu_Q
                        -M_{T_{i,m}}q_j
                        +M_{T_{i,m}}\mu_Q
                    }{
                        m \sigma_Q \Sigma_{T_{i,m}}
                    }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    \sum \limits _{0 \leq {j} \lt m}
                        q_{j}t_{i+j}
                    -
                    \sum \limits _{0 \leq {j} \lt m}
                        t_{i+j}\mu_Q
                    -
                    \sum \limits _{0 \leq {j} \lt m}
                        \left(
                            M_{T_{i,m}}q_j
                            -M_{T_{i,m}}{\mu_Q}
                        \right)
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    Q\cdot{T_{i,m}}
                    -
                    \sum \limits _{0 \leq {j} \lt m}
                        t_{i+j}\mu_Q
                    -
                    M_{T_{i,m}}
                    \sum \limits _{0 \leq {j} \lt m}
                        \left(
                            q_j-{\mu_Q}
                        \right)
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    Q\cdot{T_{i,m}}
                    -
                    \mu_Q
                    \sum \limits _{0 \leq {j} \lt m}
                        t_{i+j}
                    -
                    0
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    Q\cdot{T_{i,m}}
                    -
                    \mu_Q
                    \sum \limits _{0 \leq {j} \lt m}
                        t_{i+j}
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    Q\cdot{T_{i,m}}
                    -
                    \mu_{Q}m
                    \sum \limits _{0 \leq {j} \lt m}
                        \frac{t_{i+j}}{m}
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{
                    Q\cdot{T_{i,m}}
                    -
                    m\mu_{Q}M_{T_{i,m}}
                }{
                    m \sigma_Q \Sigma_{T_{i,m}}
                }
            \right]
        }
        \\        
\end{align}
$

# Computing the  Z-Normalized Euclidean Distance from the Pearson Correlation

Based on the fact that the Pearson Correlation, $\rho$, can be written as (see Equation 4 in [this paper](https://www.cs.unm.edu/~mueen/Projects/JOCOR/joinICDM.pdf) or Equation 3 in [this paper](https://arxiv.org/pdf/1601.02213.pdf)):

$
\begin{align}
    \rho(Q, T_{i,m}) &= \frac{E
        \left[
            \left(
                Q-\mu_Q
            \right)
            \left(
                T_{i,m}-M_{T_{i,m}}
            \right)
        \right]
    }{\sigma_{Q}\Sigma_{T_{i,m}}}
    \\
    &= 
        \frac{
            \langle
            \left(
                Q-\mu_Q
            \right)
            ,
            \left(
                T_{i,m}-M_{T_{i,m}}
            \right)
            \rangle
        }{\sigma_{Q}\Sigma_{T_{i,m}}}
    \\
    &=
        \frac{1}{m}
        \sum \limits _{0 \leq j \lt m}
            \frac{
                \left(
                    q_j-\mu_Q
                \right)
                \left(
                    t_{i+j}-M_{T_{i,m}}
                \right)
            }{\sigma_{Q}\Sigma_{T_{i,m}}}
    \\
    &=
        \frac{1}{m}
        \sum \limits _{0 \leq j \lt m}
            \left(
                \frac{
                    q_j-\mu_Q
                }{\sigma_{Q}}
            \right)
            \left(
                \frac{    
                    t_{i+j}-M_{T_{i,m}}
                }{\Sigma_{T_{i,m}}}
            \right)
    \\
\end{align}
$

# A Review of Welford's Paper

In 1962, B.P. Welford published a paper titled [Note on a Method for Calculating Corrected Sums of Squares and Products](http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.302.7503&rep=rep1&type=pdf). Let's review some important identities/derivations from this paper as they will come in handy later. Note that we use zero-based indexing when referencing elements of our time series.

### Deriving Equation (1)

$
\begin{align}
    \mu_Q^{(n)} &= 
        \sum \limits _{0 \leq j \lt n} \frac{q_j}{n}
    \\
    &=
        \frac{q_0}{n} + \frac{Q_1}{n} + \ldots + \frac{q_{n-2}}{n} + \frac{q_{n-1}}{n}
    \\
    &=
        1 \cdot \frac{q_0}{n} + 1 \cdot \frac{Q_1}{n} + \ldots + 1 \cdot \frac{q_{n-2}}{n} + \frac{q_{n-1}}{n}
    \\
    &=
        \frac{n-1}{n-1} \frac{q_0}{n} + \frac{n-1}{n-1} \frac{q_1}{n} + \ldots + \frac{n-1}{n-1} \frac{q_{n-2}}{n} + \frac{q_{n-1}}{n}
    \\
    &=
        \frac{n-1}{n} \frac{q_0}{n-1} + \frac{n-1}{n} \frac{q_1}{n-1} + \ldots + \frac{n-1}{n} \frac{q_{n-2}}{n} + \frac{q_{n-1}}{n}
    \\
    &=
        \frac{n-1}{n}
        \left(
            \frac{q_0}{n-1} + \frac{q_1}{n-1} + \ldots + \frac{q_{n-2}}{n-1}
        \right)
        + \frac{q_{n-1}}{n}
    \\
    &=
        \frac{n-1}{n}
        \sum \limits _{0 \leq j \lt n-1} \frac{q_j}{n-1}
        + \frac{q_{n-1}}{n}
    \\
    &=
        \frac{n-1}{n}
        \mu_Q^{(n-1)}
        + \frac{q_{n-1}}{n}
    \\
    \mu^{(m)} &=
        \frac{m-1}{m}
        \mu_Q^{(m-1)}
        + \frac{q_{m-1}}{m}
    \\
\end{align}
$

### Deriving Equation (2)

$
\begin{align}
    i < n-1, \;\;\;\;\;
    q_i - \mu_Q^{(n)} &= 
        q_i - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        - \frac{q_{n-1}}{n}
    \\
    &=
        q_i 
        - \frac{n}{n} \mu_Q^{(n-1)}
        + \frac{\mu_Q^{(n-1)}}{n}
        - \frac{q_{n-1}}{n}
    \\
    &=
        q_i 
        - \mu_Q^{(n-1)}
        + \frac{1}{n}
        \left(
        -q_{n-1} + \mu_Q^{(n-1)}
        \right)
    \\
    &=
        q_i 
        - \mu_Q^{(n-1)}
        - \frac{1}{n}
        \left(
        q_{n-1} - \mu_Q^{(n-1)}
        \right)
    \\
    q_i - \mu_Q^{(m)} &= 
        q_i 
        - \mu_Q^{(m-1)}
        - \frac{1}{m}
        \left(
        q_{m-1} - \mu_Q^{(m-1)}
        \right)
    \\
\end{align}
$

### Deriving Equation (3)

$
\begin{align}
    i = n-1, \;\;\;\;\;
    q_{n-1} - \mu_Q^{(n)} &= 
        q_{n-1} - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        - \frac{q_{n-1}}{n}
    \\
    &=
        - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        - \frac{q_{n-1}}{n} + q_{n-1}
    \\
    &=
        - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        + q_{n-1}
        \left(
        -\frac{1}{n} + 1
        \right)
    \\
    &=
        - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        + q_{n-1}
        \left(
        -\frac{1}{n} + \frac{n}{n}
        \right)
    \\
    &=
        - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        + q_{n-1}
        \left(
        -\frac{1}{n} + \frac{n}{n}
        \right)
    \\
    &=
        - \frac{n-1}{n}
        \mu_Q^{(n-1)}
        + q_{n-1}
        \left(
        \frac{n-1}{n}
        \right)
    \\
    &=
        \frac{n-1}{n}
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right)
    \\
    q_{m-1} - \mu_Q^{(m)} &= \frac{m-1}{m}
        \left(
            q_{m-1}
            - 
            \mu_Q^{(m-1)}
        \right)
\end{align}
$

### Deriving Equation (I) - The Corrected Sum of Squares

for $n = 1, 2, \ldots, m$

$
\begin{align}
    S^{(n)} &= 
        \sum \limits _{0 \leq i \lt n}
            \left(
                q_i - \mu_Q^{(n)}
            \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
                (q_i - \mu_Q^{(n)})^2
        +
        \left(
            q_{n-1} - \mu_Q^{(n)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left[
                q_i
                - \mu_Q^{(n-1)}
                - \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
            \right] ^2
        +
        \left[
            \frac{n-1}{n}
            \left(
                q_{n-1}
                - 
                \mu_Q^{(n-1)}
            \right)
        \right] ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt m-1}
            \left[
                \left(
                    q_i 
                    - \mu_Q^{(n-1)}
                \right)
                - \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
            \right] ^2
        +
        \left(
            \frac{n-1}{n}
        \right) ^2
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left[
                \left(
                    q_i 
                    - \mu_Q^{(n-1)}
                \right) ^2
                - \frac{1}{n^2}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right) ^2
            \right]
        +
        \frac{(n-1)^2}{n^2}
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left(
                q_i 
                - \mu_Q^{(n-1)}
            \right) ^2
        - 
        \sum \limits _{0 \leq i \lt n-1}
            \frac{1}{n^2}
            \left(
                q_{n-1} - \mu_Q^{(n-1)}
            \right) ^2
        +
        \frac{(n-1)^2}{n^2}
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left(
                q_i 
                - \mu_Q^{(n-1)}
            \right) ^2
        - 
        \frac{n-1}{n^2}
        \left(
            q_{n-1} - \mu_Q^{(n-1)}
        \right) ^2
        +
        \frac{(n-1)^2}{n^2}
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left(
                q_i 
                - \mu_Q^{(n-1)}
            \right) ^2
        -
        \left[
            \frac{n-1}{n^2}
            +
            \frac{(n-1)^2}{n^2}
        \right]
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left(
                q_i 
                - \mu_Q^{(n-1)}
            \right) ^2
        -
        \left[
            \frac{n-1+n^2-2n+1}{n^2}
        \right]
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        \sum \limits _{0 \leq i \lt n-1}
            \left(
                q_i 
                - \mu_Q^{(n-1)}
            \right) ^2
        -
        \left[
            \frac{n^2-n}{n^2}
        \right]
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    &=
        S^{(n-1)}
        +
        \left(
            \frac{n-1}{n}
        \right)
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right) ^2
    \\
    S^{(m)} &=
        S^{(m-1)}
        +
        \left(
            \frac{m-1}{m}
        \right)
        \left(
            q_{m-1}
            - 
            \mu_Q^{(m-1)}
        \right) ^2
    \\
\end{align}
$

### Deriving Equation (II) - The Corrected Sum of Products

$
\begin{align}
    S^{(n)} &= 
        \sum \limits _{0 \leq j \lt n}
            \left(
                q_j - \mu_Q^{(n)}
            \right)
            \left(
                t_{i+j} - T_{i,n}
            \right)
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_j - \mu_Q^{(n)}
            \right)
            \left(
                t_{i+j} - T_{i,n}
            \right)
        +
        \left(
            q_{n-1} - \mu_Q^{(n)}
        \right)
        \left(
            t_{i+n-1} - T_{i,n}
        \right)
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left[
                q_j
                - \mu_Q^{(n-1)}
                - \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
            \right]
            \left[
                t_{i+j}
                - T_{i,n-1}
                - \frac{1}{n}
                \left(
                    t_{i+n-1} - T_{i,n-1}
                \right)
            \right]
        +
        \left[
            \frac{n-1}{n}
            \left(
                q_{n-1}
                - 
                \mu_Q^{(n-1)}
            \right)
        \right]
        \left[
            \frac{n-1}{n}
            \left(
                t_{i+n-1}
                - 
                T_{i,n-1}
            \right)
        \right]
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left[
                \left(
                    q_j
                    - \mu_Q^{(n-1)}
                \right)
                - \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
            \right]
            \left[
                \left(
                    t_{i+j}
                    - T_{i,n-1}
                \right)
                - \frac{1}{n}
                \left(
                    t_{i+n-1} - T_{i,n-1}
                \right)
            \right]
        +
        \left(
            \frac{n-1}{n}
        \right)^2
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right)
        \left(
            t_{i+n-1}
            - 
            T_{i,n-1}
        \right)
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left[
                \left(
                    q_j
                    - \mu_Q^{(n-1)}
                \right)
                \left(
                    t_{i+j}
                    - T_{i,n-1}
                \right)
                -
                \left(
                    q_j
                    - \mu_Q^{(n-1)}
                \right)
                \frac{1}{n}
                \left(
                    t_{i+n-1} - T_{i,n-1}
                \right)
                - \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
                \left(
                    t_{i+j}
                    - T_{i,n-1}
                \right)
                +
                \frac{1}{n}
                \left(
                    q_{n-1} - \mu_Q^{(n-1)}
                \right)
                \frac{1}{n}
                \left(
                    t_{i+n-1} - T_{i,n-1}
                \right)
            \right]
        +
        \left(
            \frac{n-1}{n}
        \right)^2
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right)
        \left(
            t_{i+n-1}
            - 
            T_{i,n-1}
        \right)
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_j
                - \mu_Q^{(n-1)}
            \right)
            \left(
                t_{i+j}
                - T_{i,n-1}
            \right)
        -\frac{1}{n}
        \left(
            t_{i+n-1} - T_{i,n-1}
        \right)
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_j
                - \mu_Q^{(n-1)}
            \right)
        - \frac{1}{n}
        \left(
            q_{n-1} - \mu_Q^{(n-1)}
        \right)
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                t_{i+j}
                - T_{i,n-1}
            \right)
        +\frac{1}{n^2}
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_{n-1} - \mu_Q^{(n-1)}
            \right)
            \left(
                t_{i+n-1} - T_{i,n-1}
            \right)
        +
        \left(
            \frac{n-1}{n}
        \right)^2
        \left(
            q_{n-1}
            - 
            \mu_Q^{(n-1)}
        \right)
        \left(
            t_{i+n-1}
            - 
            T_{i,n-1}
        \right)
    \\
    &= 
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_j
                - \mu_Q^{(n-1)}
            \right)
            \left(
                t_{i+j}
                - T_{i,n-1}
            \right)
        -\frac{1}{n}
        \left(
            t_{i+n-1} - T_{i,n-1}
        \right)
        \sum \limits _{0 \leq j \lt n-1}
            \left(
                q_j
                - \mu_Q^{(n-1)}
            \right)
        - \frac{1}{n}
        \left(
            q_{n-1} - \mu_Q^{(n-1)}
        \right)

# Covariance

Recall that the covariance, $cov(Q,T_{i,m})$, can be written as:

$
\begin{align}
    cov(Q,T_{i,m}) &= E
        \left[
            \left(
                Q-\mu_Q^{(m)}
            \right)
            \left(
                T_{i,m}-M_{T_{i,m}}
            \right)
        \right]
    \\
    &= 
        \langle
            \left(
                Q-\mu_Q^{(m)}
            \right)
            ,
            \left(
                T_{i,m}-M_{T_{i,m}}
            \right)
        \rangle
    \\
    &=
        \frac{1}{m}
        \sum \limits _{0 \leq j \lt m}
            \left(
                q_j-\mu_Q^{(m)}
            \right)
            \left(
                t_{i+j}-M_{T_{i,m}}
            \right)
    \\
\end{align}
$

Note that we've explicitly called out the fact that the means, $\mu_Q^{(m)}$ and $M_{T_{i,m}}$, are computed with the subsequences of length $m$. Additionally, according to Welford, we can express these means with respect to the means of the same subsequences that have their last elements removed (i.e., , $\mu_Q^{(m-1)}$ and $M_{T_{i,m-1}}$).

$
\begin{align}
    cov(Q,T_{i,m}) 
    &=
        \frac{1}{m}
        \sum \limits _{0 \leq j \lt m}
            \left(
                q_j-\mu_Q^{(m)}
            \right)
            \left(
                t_{i+j}-M_{T_{i,m}}
            \right)
    \\
    &=
        \frac{1}{m}
        \sum \limits _{0 \leq j \lt m}
            \left(
                q_j-\mu_Q^{(m-1)}+\frac{q_{m-1}-\mu_Q^{(m-1)}}{m}
            \right)
            \left(
                t_{i+j}-M_{T_{i,m-1}}+\frac{t_{i+m-1}-M_{T_{i,m-1}}}{m}
            \right)
\end{align}
$

In contrast, the covariance for the the same pair of subsequences with the last elements removed would be:

$
\begin{align}
    cov(Q_{m-1},T_{i,m-1}) &= 
        \frac{1}{m-1}
        \sum \limits _{0 \leq j \lt m-1}
            \left(
                q_j-\mu_Q^{(m-1)}
            \right)
            \left(
                t_{i+j}-M_{T_{i,m-1}}
            \right)
    \\
\end{align}
$

Similar to above, the Z-normalized Euclidean distance can be computed from $\rho$ following:

$
\begin{align}
    D(Q, T_{i,m}) &=
        \sqrt{
            \sum \limits _{0 \leq {j} \lt m}
                \left(
                    \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    - 
                    \frac{q_j-\mu_Q}{\sigma_Q}
                \right)^2
        }
        \\
    \vdots
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \frac{1}{m}
                \sum \limits _{0 \leq {j} \lt m}
                    \left(
                        \frac{t_{i+j}-M_{T_{i,m}}}{\Sigma_{T_{i,m}}}
                    \right)
                    \left(
                        \frac{q_j-\mu_Q}{\sigma_Q}
                    \right)
            \right]
        }
        \\
    &=
        \sqrt{
            2m
            \left[
                1
                -
                \rho(Q,T_{i,m})
            \right]
        }
        \\
\end{align}
$
